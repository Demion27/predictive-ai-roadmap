```python
import import_ipynb
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from stock_simulator import wiener_process
```

$S_{t} = S_{o}\exp{(μ-\frac{σ}{2}+σW)}$

```python
def gbm(n_steps = 252, T=1, n_paths = 5,mu = 0.1, sigma = 0.2,S0 = 100):
    '''Simulate Geometric Brownian Motion paths.
        param n_steps: Number of time steps
        param T: Total time
        param n_paths: Number of paths to simulate  
        param mu: Drift coefficient
        param sigma: Volatility coefficient
        param S0: Initial stock price
        return: Simulated GBM paths as a pandas DataFrame with time column and path columns
    '''
    dt = T/n_steps
    t = np.arange(0, T, dt).reshape(n_steps,1)
    W,times = wiener_process(n_steps=n_steps, n_paths=n_paths)
    S = np.full(shape = n_paths, fill_value =S0).reshape(1,n_paths) * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)
    
    # Create DataFrame with time and path columns
    df = pd.DataFrame(S, columns=[f'Path_{i}' for i in range(n_paths)])
    df.insert(0, 'time', times)
    return df
```

```python
plt.figure(figsize=(10,6))
gbm_paths = gbm(n_paths=5)
gbm_paths.plot(x='time', y=[f'Path_{i}' for i in range(5)], label=[f'GBM Path {i}' for i in range(5)])
plt.title('Simulated Geometric Brownian Motion Paths')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.grid()
plt.show()
```
